# 3.1. Combining the datasets into one DataFrame


```python
# Drop the 'Unnamed' column in the tmdb_movies dataset
tmdb_movies.drop('Unnamed: 0', axis=1, inplace=True)
```

```python

# Assign the 'data' DataFrame to the Datasets
df = pd.DataFrame()

data0 = pd.read_csv("tmdb.movies.csv")
data1 = pd.read_csv("tn.movie_budgets.csv")
data2 = pd.read_csv("bom.movie_gross.csv")

tn_movie_budgets = data1
bom_movies = data2
tmdb_movies = data0
# Combining the 'bom-movies' and the 'tmdb_movies' datasets on 'title column

combined_data = pd.merge(bom_movies, tmdb_movies, on=['title'], how='outer')
 
# Combining the resulting data above with 'tn_movie_budgets' dataset
df_combined = pd.concat([combined_data, tn_movie_budgets], join='outer')

df_combined['domestic_gross'] = df_combined['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)

df_combined.head()
```

```python
# Finding and handling missing values appropriately
df_combined.isna().sum()
```

```python
df_combined =df_combined.drop(['Unnamed: 0', 'genre_ids', 'id', 'original_language'], axis=1)
```

# 3.2. Formatting Datatypes 

```python
# Converting the  release_date to datetime
df_combined['release_date'] = pd.to_datetime(df_combined['release_date'], errors='coerce')  # coerce converts invalid dates to NaT
# Converting 'domestic_gross' to float
df_combined['domestic_gross'] = df_combined['domestic_gross'].astype(float)

# Converting 'foreign_gross' to float
df_combined['foreign_gross'] = pd.to_numeric(df_combined['foreign_gross'], errors='coerce')  # coerce converts non-numeric values to NaN

# Converting 'year' to integer (assuming it's a year)
df_combined['year'] = df_combined['year'].astype('Int64')  # Use 'Int64' for nullable integer

# Converting 'release_date' to datetime


# Converting 'production_budget' and 'worldwide_gross' to float
df_combined['production_budget'] = pd.to_numeric(df_combined['production_budget'], errors='coerce')
df_combined['worldwide_gross'] = pd.to_numeric(df_combined['worldwide_gross'], errors='coerce')

df_combined.head()
```

# 3.3. Data Wrangling

```python
# Fill NAN values in 'release_date' column with mean
df_combined['release_date'] = df_combined['release_date'].fillna('2000-01-01')
df_combined['release_date'] = pd.to_datetime(df_combined['release_date'])
# Fill NAN values in 'vote_count' and 'vote_average' columns with mean
df_combined['vote_count'] = df_combined['vote_count'].fillna(df_combined['vote_count'].mean())
df_combined['vote_average'] = df_combined['vote_average'].fillna(df_combined['vote_average'].mean())
# Fill NAN values in 'domestic_gross', 'worldwide_gross' and 'foreign_gross' columns with 0
df_combined['domestic_gross'].fillna(0, inplace=True)
df_combined['worldwide_gross'].fillna(0, inplace=True)
df_combined['foreign_gross'].fillna(0, inplace=True)
# Drop NAN values in 'original_title' column
df_combined = df_combined.dropna(subset=['original_title'])
# Fill NAN values in 'studio' and 'movie' column with 'Uknown'
df_combined['movie'] = df_combined['movie'].fillna('Unknown')
df_combined['studio'] = df_combined['studio'].fillna('Unknown')
# Fill NAN values in 'production_budget' column with 0
df_combined['production_budget'] = df_combined['production_budget'].fillna(0)

# Sort the data by 'release_date' in ascending order
df_sorted = df_combined.sort_values(by='release_date', ascending=True)

# Reset the index
df_sorted = df_sorted.reset_index(drop=True)

df_sorted.head()
```

# 4.0.  Exploratory Data Analysis
In this section we have to explore the data, perform calculations involving statistical methods (e.g,mean), finding the correlation between two variables and finally visualizing the data output.

```python
# Showing the top ten movies according to domestic_gross
df_combined['domestic_gross'] = df_combined['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)

# Sort the DataFrame by 'domestic_gross' in descending order
top_ten_movies = df_combined.sort_values(by='domestic_gross', ascending=False).head(10)

print(top_ten_movies[['title', 'domestic_gross']])
```

```python
# Visualizing the top_ten_movies
plt.figure(figsize=(12, 6))
plt.bar(top_ten_movies['title'], top_ten_movies['domestic_gross'], color='magenta')
plt.title('Top Ten Movies by Domestic Gross')
plt.xlabel('Movie Title')
plt.ylabel('Domestic Gross ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4.1. Statistical Methods

Here we use the cleaned data to calculate measures of central tendency, measures of dispersion, correlation and visualize the output clearly.

This helps in finding the relationship between various variables and probably guide in determining the be actions to take on the output.

```python
# Measures of central tendency(Mean)
central_tendency = df_combined[['domestic_gross', 'foreign_gross', 'popularity', 'vote_average', 'vote_count']].mean()
print("Mean:", central_tendency)

# Measures of Dispersion(Standard Deviation)
std = df_combined[['domestic_gross', 'foreign_gross', 'popularity', 'vote_average', 'vote_count']].std()
print("Standard Deviation:", std)

# Calculating Interquartile Range (IQR) for Middle Fifty
IQR = df_combined[['domestic_gross', 'foreign_gross', 'popularity', 'vote_average', 'vote_count']].quantile([0.25, 0.75])
print("Interquartile Range (IQR):", IQR)


# Calculating Correlation Matrix
correlation_matrix = df_combined[['domestic_gross', 'foreign_gross', 'popularity', 'vote_average', 'vote_count']].corr()
print("Correlation Matrix:", correlation_matrix)
```
